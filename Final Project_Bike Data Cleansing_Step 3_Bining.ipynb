```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pprint import pprint
from datetime import datetime
```

```python
NYC_Trip_201801 = pd.read_csv("./Resources/Bike_Data/NYC_Citibike/2018/Step_2/NYC_Trip_201801.csv")
NYC_Trip_201801
```

```python
NYC_Trip_201801=NYC_Trip_201801.set_index("start_station_name")

# NYC_Trip_201801.groupby(["start_station_name"])
```

```python
NYC_Trip_201801_analysis = NYC_Trip_201801.groupby(["start_date", "start_station_name", "start_hour"])["trip_duration"].mean()
NYC_Trip_201801_analysis_count = NYC_Trip_201801.groupby(["start_date", "start_station_name", "start_hour"])["trip_duration"].count()

NYC_Trip_201801_analysis = pd.DataFrame(NYC_Trip_201801_analysis).reset_index()
NYC_Trip_201801_analysis.head()
```

```python
NYC_Trip_201801_analysis_count = pd.DataFrame(NYC_Trip_201801_analysis_count).reset_index()
# NYC_Trip_201801_analysis_count
NYC_Trip_201801_analysis['trip_count'] = NYC_Trip_201801_analysis_count['trip_duration']
NYC_Trip_201801_analysis = NYC_Trip_201801_analysis.rename(columns={"trip_duration": "avg_trip_duration"})
NYC_Trip_201801_analysis['avg_trip_duration'].round()
NYC_Trip_201801_analysis.head()
```
